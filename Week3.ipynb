```python
import folium
map1= folium.Map(location=[43.6532, -79.3832], zoom_start=11)

cl = {'North York':'red','Downtown Toronto':'blue','Scarborough':'green','Etobicoke': 'purple','Central Toronto':'orange','West Toronto':'darkpurple','York': 'darkpurple','East York':'darkred','East Toronto':'gray','Mississauga':'black'}
for lat, lng, label,clrd in zip(df2['Latitude'], df2['Longitude'], df2['Neighborhood'], df2['Borough']):
    clr = cl[clrd]
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color=clr,
        fill=True,
        fill_color=clr,
        fill_opacity=0.7,
        parse_html=False).add_to(map1)  
    
```
